```python
# Para cada arquivo na subpasta ./dados/csv, ler o arquivo e concatenar em um único DataFrame
import os
import pandas as pd

caminho_dados = "../dados/csv"
lista_arquivos = os.listdir(caminho_dados)
lista_arquivos = [arquivo for arquivo in lista_arquivos if arquivo.endswith(".csv")]
df_list = []

for arquivo in lista_arquivos:
    caminho_arquivo = os.path.join(caminho_dados, arquivo)
    # Tentar diferentes encodings
    try:
        df_temp = pd.read_csv(caminho_arquivo, sep=',', encoding='utf-8')
    except UnicodeDecodeError:
        try:
            df_temp = pd.read_csv(caminho_arquivo, sep=',', encoding='latin-1')
        except UnicodeDecodeError:
            df_temp = pd.read_csv(caminho_arquivo, sep=',', encoding='iso-8859-1')
    df_list.append(df_temp)

df = pd.concat(df_list, ignore_index=True)
df.shape
```

```python
# ordenar por Numero Scheda PS
df.sort_values(by='Numero Scheda PS', ascending=False).head()
```

```python
# contar quantas linhas duplicadas existem no df
print(df.duplicated().sum())

# Criar um dataframe somente com as linhas duplicadas
df_duplicados = df[df.duplicated(keep=False)]
df_duplicados.sort_values(by='Numero Scheda PS', ascending=False).head()
```

```python
# Apagar a primeira linha de cada duplicado
df_unicos = df.drop_duplicates(keep='first')
df_unicos.sort_values(by='Numero Scheda PS', ascending=False).head()
```
